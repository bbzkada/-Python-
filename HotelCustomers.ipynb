*探索分销渠道与预订结果的关联:比较通过不同分销渠道预订的客户在取消率、未出现率和回头客方面的差异。*


前项为DistributionChannel、MarketSegment。


后项为BookingsCanceled、BookingsNoShowed、BookingsCheckedIn。

```python
#选择我们需要的属性（对应过滤器）
data = data[['ID',#标识字段
         'DistributionChannel', 'MarketSegment',#前项
         'BookingsCanceled','BookingsNoShowed','BookingsCheckedIn']]#后项

# 打印DataFrame的前几行进行检查
data.head(10)
```

对BookingsCanceled、BookingsNoShowed、BookingsCheckedIn的值进行重新分组

```python
#对BookingsCanceled、BookingsNoShowed、BookingsCheckedIn进行二元化（非0即1）
def to_brinary(value):
    if value == 0:
        return False
    else:
        return True

data['BookingsCanceled'] = data['BookingsCanceled'].apply(to_brinary)
data['BookingsNoShowed'] = data['BookingsNoShowed'].apply(to_brinary)
data['BookingsCheckedIn'] = data['BookingsCheckedIn'].apply(to_brinary)

data.head(10)
```

```python
# 对前项进行独热编码
encoded_data = pd.get_dummies(data[['DistributionChannel', 'MarketSegment']])

# 选取后项列进行独热编码
back_items = pd.get_dummies(data[['BookingsCanceled', 'BookingsNoShowed', 'BookingsCheckedIn']])

# 合并编码后的前项和后项
combined_data = pd.concat([encoded_data, back_items], axis=1)

# 使用 Apriori 算法计算频繁项集
frequent_itemsets = apriori(combined_data, min_support=0.1, use_colnames=True)

# 生成关联规则
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# 筛选出前项为 DistributionChannel 和 MarketSegment 独热编码后的列，后项为指定后项列的规则
desired_rules = rules[(rules['antecedents'].apply(lambda x: all(col.split('_')[0] in ['DistributionChannel', 'MarketSegment'] for col in x)) &
                       rules['consequents'].apply(lambda x: all(col in back_items.columns for col in x)))]

desired_rules.head(10)
```
